## Basic Signal Processing

```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Enshu1
sin波とcos波を1つのfigureで表示

```python
n = 8
f = np.arange(0, 1, 1/n)
x = f * 2 * np.pi

#sin波
y_sin = np.sin(x)

#cos波
y_cos = np.cos(x)
```

```python
# グラフに出力してみる
fig = plt.figure()
sp1 = fig.add_subplot(2, 1, 1)
sp1.plot(x,y_sin)
sp1.set_xticks(x)


sp2 = fig.add_subplot(2, 1, 2)
sp2.plot(x,y_cos)
sp2.set_xticks(x)

plt.show()
```

## Enshu2
sin波とcos波を1つのfigで合成

```python
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.show()
```

```python
_, ax = plt.subplots()
ax.plot(x,y_sin)
ax.plot(x,y_cos)
ax.set_xticks(x)
plt.show()
```

## Enshu3
sin波とcos波を3Dで表示

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, x, y_sin)
ax.plot(x, x+1, y_cos)
ax.plot(x, x+2, y_sin)
ax.plot(x, x+3, y_cos)
ax.plot(x, x+4, y_sin)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i,z in enumerate([y_sin, y_cos, y_sin, y_cos, y_sin]):
    ax.plot(x,np.full(len(x),i*0.1),z)

plt.savefig('fig.png')
```

## ファイルへ書き出し

```python
with open('sample.txt', 'w') as f:
    f.write('sin,cos\n')
    f.write('0,0.1\n')
    for line in np.array([y_sin,y_cos]).T:
        f.write(f'{",".join(map(lambda x: str(round(x,2)), line))}\n')
```
